```python
data['age_years'] = data['age'] // 365

data['age_group'] = pd.cut(data['age_years'], 
                         bins=[0, 29, 39, 49, 59, 69, 120],
                         labels=['20s','30s','40s','50s','60s','70+'])

data
```

<h3>BMI Calculator</h3>

```python
BmiData = data.copy()
data['BMI'] = data['weight'] / ((data['height'] / 100) ** 2)

data['BMI_group'] = pd.cut(
    BmiData['BMI'],
    bins=[0, 18.5, 25, 30, 100],
    labels=['Underweight', 'Normal', 'Overweight', 'Obese']
)

data.head(6)
```

<h3>correlation of dataset</h3>

```python
data2 = data.drop(['age_group'], axis=1)
data_numeric = data2.select_dtypes(include=['number'])
correlation = data_numeric.corr()
correlation
```
